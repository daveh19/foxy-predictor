```python
import pandas as pd
import io
from bs4 import BeautifulSoup
import urllib.request
import json
```

```python
wahlrecht = 'http://www.wahlrecht.de/umfragen/allensbach.htm'
git_emnid = 'https://gitlab.tubit.tu-berlin.de/moboehle/Predictor/blob/master/data/emnid.csv'
```

```python
page = urllib.request.urlopen(wahlrecht)
soup = BeautifulSoup(page, 'html.parser')

head = soup.find('thead')
body = soup.find('tbody')

table = []
#idx = []
rows = body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    #idx.append(cols[0])
    table.append([ele for ele in cols if ele]) 

header = []
cols = head.find_all('th')
for col in cols:
    if col.get_text() != '\xa0':
        header.append(col.get_text())
header.insert(0, 'Date')


df = pd.DataFrame(table, columns=header)
```
