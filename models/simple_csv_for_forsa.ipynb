```python
import numpy as  np
from  statsmodels.nonparametric import smoothers_lowess
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=np.genfromtxt('modified_forsa.csv',  delimiter=',') #importing the csv file forsa
```

```python
results=data[:,1:7] # only taking into account the values for the 6 parties
```

```python
results=results.T
```

```python
[plt.plot(results[i,:]) for i in range (6) ]; # plotting the polls over time
```

```python
covar=np.cov(results) #calculating the covariance matrix over the 6 parties
```

```python
import statsmodels.nonparametric.smoothers_lowess as stat #smoothing the data using local regression
smoothed = np.zeros((6,185,2))
smoothed = [*(map(lambda i : stat.lowess(results[i,:], range(185) , frac=0.1), range (6)))]  #frac determines the smoothing
```

```python
[plt.plot(range(185), smoothed[i][:,1]) for i in range (6)];
```
