```python
from __future__ import division, print_function
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
parties = ['CDU', 'SPD', 'Gruene', 'FDP', 'Linke', 'AfD', 'Sonstige']
```

```python
# Optional: Read one of the original CSV files and convert it to the cleaned-up pandas format.
def p2f(x):
    return float(x.strip('%'))/100

data = pd.read_csv('../data/allensbach-original.csv', sep=';', converters={party: p2f for party in parties})

data = data.drop(data.columns[1], axis=1)
data = data.drop(data.columns[8], axis=1)
#data = data.drop(data.columns[-1], axis=1)

#data.to_csv('../data/allensbach.csv', index=False)
```

```python
# Read data file.
data = pd.read_csv('../data/emnid.csv')
data
```

```python
all_results = data[parties].as_matrix()
all_results.shape
```

```python
# Split the data up: Use polling results from 5 consecutive time points as input and predict the next time point. 
timepoints_per_input = 5

X = []
y = []

for start in range(0, len(all_results)-timepoints_per_input, timepoints_per_input+1):
    X.append(all_results[start:start+timepoints_per_input].flatten())
    y.append(all_results[start+timepoints_per_input])
    
X = np.array(X)
y = np.array(y)
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
model.score(X_test, y_test)
```
